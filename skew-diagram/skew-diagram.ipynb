```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def readFASTAs(fileName):

    with open(fileName, 'r') as file:
        v = []
        genome = ''
        for line in file:
            if line[0] != '>':
                genome += line.strip()
            else:
                v.append(genome.upper())
                genome = ''
        v.append(genome.upper())
        del v[0]
        return v
```

```python
Sequences = readFASTAs('negative591_131bp.txt')
Sequences = Sequences[:400]
len(Sequences)
```

```python
for sequence in Sequences: 
    c = 0
    v = []
    v.append(c)

    for bp in sequence:
        if bp == 'C':
            c -= 1
            v.append(c)
        elif bp == 'G':
            c += 1
            v.append(c)
        else:
            c += 0
            v.append(c)
    plt.plot(v)
    plt.ylim(-30, 40)
plt.savefig(dpi=300, fname = 'negative')
plt.show()
```
